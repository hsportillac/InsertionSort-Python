## Sobre todas las permutaciones usando un unico vector

```python
import numpy as np
import matplotlib.pyplot as plt

n = 8
stv = []
comp = []
ques = []
intc = []

def isortSteps(v):
    vaux = []
    for i in range(len(v)):
        vaux.append(v[i])
    steps = 0
    for i in range(1,len(vaux)):
        x = vaux[i]
        j = i-1
        while (j > -1) and (vaux[j] > x):
            vaux[j+1] = vaux[j]
            j = j -1
            steps = steps + 3
        steps = steps + 1
        vaux[j+1] = x
        steps = steps + 4
    steps = steps + 1
    return steps

def isortStepsComp(v):
    vaux = []
    for i in range(len(v)):
        vaux.append(v[i])
    steps = 0
    for i in range(1,len(vaux)):
        x = vaux[i]
        j = i-1
        while (j > -1) and (vaux[j] > x):
            vaux[j+1] = vaux[j]
            j = j -1
            steps = steps + 1
        if (j != -1):
            steps = steps + 1
        vaux[j+1] = x
    return steps

def isortStepsQu(v):
    vaux = []
    for i in range(len(v)):
        vaux.append(v[i])
    steps = 0
    for i in range(1,len(vaux)):
        x = vaux[i]
        j = i-1
        while (j > -1) and (vaux[j] > x):
            vaux[j+1] = vaux[j]
            j = j -1
            steps = steps + 1
        steps = steps + 1
        vaux[j+1] = x
    return steps

def isortStepsIn(v):
    vaux = []
    for i in range(len(v)):
        vaux.append(v[i])
    steps = 0
    for i in range(1,len(vaux)):
        x = vaux[i]
        j = i-1
        while (j > -1) and (vaux[j] > x):
            vaux[j+1] = vaux[j]
            j = j -1
            steps = steps + 1
        vaux[j+1] = x
    return steps

def createVector(n):
    v = []
    for i in range (0,n):
        v.append(i+1)
    return v    

def swap (v, i, j):
    t = v[i];
    v[i] = v[j];
    v[j] = t;
    
def perm(v, i, n):
    j = 0;
    k = 0;
    l = 0;
    if (i == n):
        stv.append(isortSteps(v))
        comp.append(isortStepsComp(v))
        ques.append(isortStepsQu(v))
        intc.append(isortStepsIn(v))
    else:
        for j in range(i,n):
            swap(v,i,j)
            perm(v,i+1,n)
            swap(v,i,j)

v = createVector(n)
perm(v,0,n)

print'EVALUACION DE INSTRUCCIONES'
print 'Mejor No de instrucciones teorico: ' + str(5*n - 4) 
print 'Promedio de No de instrucciones teorico: ' + str((3.0/4.0)*n**2 + (17.0/4.0)*n - 4)
print 'Peor No de instrucciones teorico: ' + str((3.0/2.0)*n**2 + (7.0/2.0)*n - 4)
print 'Mejor No de instrucciones practico: ' , np.amin(stv)
print 'Promedio de No de instrucciones practico: ' , np.mean(stv)
print 'Peor No de instrucciones practico: ' , np.amax(stv)
print''
print'EVALUACION DE COMPARACIONES'
print 'Mejor No de comparaciones practico: ' , np.amin(comp)
print 'Promedio de No de comparaciones practico: ' , np.mean(comp)
print 'Peor No de comparaciones practico: ' , np.amax(comp)
print''
print'EVALUACION DE PREGUNTAS'
print 'Mejor No de preguntas practico: ' , np.amin(ques)
print 'Promedio de No de preguntas practico: ' , np.mean(ques)
print 'Peor No de preguntas practico: ' , np.amax(ques)
print''
print'EVALUACION DE INTERCAMBIOS'
print 'Mejor No de intercambios practico: ' , np.amin(intc)
print 'Promedio de No de intercambios practico: ' , np.mean(intc)
print 'Peor No de intercambios practico: ' , np.amax(intc)

plt.figure(1)
plt.subplot(221)
plt.hist(stv)
plt.title('Distribucion del numero de instrucciones')

plt.subplot(222)
plt.hist(comp)
plt.title('Distribucion del numero de comparaciones')

plt.subplot(223)
plt.hist(ques)
plt.title('Distribucion del numero de preguntas')

plt.subplot(224)
plt.hist(intc)
plt.title('Distribucion del numero de intercambios')


plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=0.35)

plt.show()

```

## Sobre todas las permutaciones usando Itertools

```python
import itertools

stv = []
comp = []
ques = []
intc = []

def isortSteps(v):
    vaux = []
    for i in range(len(v)):
        vaux.append(v[i])
    steps = 0
    for i in range(1,len(vaux)):
        x = vaux[i]
        j = i-1
        while (j > -1) and (vaux[j] > x):
            vaux[j+1] = vaux[j]
            j = j -1
            steps = steps + 3
        steps = steps + 1
        vaux[j+1] = x
        steps = steps + 4
    steps = steps + 1
    return steps

def isortStepsComp(v):
    vaux = []
    for i in range(len(v)):
        vaux.append(v[i])
    steps = 0
    for i in range(1,len(vaux)):
        x = vaux[i]
        j = i-1
        while (j > -1) and (vaux[j] > x):
            vaux[j+1] = vaux[j]
            j = j -1
            steps = steps + 1
        if (j != -1):
            steps = steps + 1
        vaux[j+1] = x
    return steps

def isortStepsQu(v):
    vaux = []
    for i in range(len(v)):
        vaux.append(v[i])
    steps = 0
    for i in range(1,len(vaux)):
        x = vaux[i]
        j = i-1
        while (j > -1) and (vaux[j] > x):
            vaux[j+1] = vaux[j]
            j = j -1
            steps = steps + 1
        steps = steps + 1
        vaux[j+1] = x
    return steps

def isortStepsIn(v):
    vaux = []
    for i in range(len(v)):
        vaux.append(v[i])
    steps = 0
    for i in range(1,len(vaux)):
        x = vaux[i]
        j = i-1
        while (j > -1) and (vaux[j] > x):
            vaux[j+1] = vaux[j]
            j = j -1
            steps = steps + 1
        vaux[j+1] = x
    return steps

def perm(v, i, n):
    v2 = list(itertools.permutations(v, n))
    for i in v2:
        stv.append(isortSteps(i))
        comp.append(isortStepsComp(i))
        ques.append(isortStepsQu(i))
        intc.append(isortStepsIn(i))
        
perm(v,0,n)

print'EVALUACION DE INSTRUCCIONES'
print 'Mejor No de instrucciones teorico: ' + str(5*n - 4) 
print 'Promedio de No de instrucciones teorico: ' + str((3.0/4.0)*n**2 + (17.0/4.0)*n - 4)
print 'Peor No de instrucciones teorico: ' + str((3.0/2.0)*n**2 + (7.0/2.0)*n - 4)
print 'Mejor No de instrucciones practico: ' , np.amin(stv)
print 'Promedio de No de instrucciones practico: ' , np.mean(stv)
print 'Peor No de instrucciones practico: ' , np.amax(stv)
print''
print'EVALUACION DE COMPARACIONES'
print 'Mejor No de comparaciones practico: ' , np.amin(comp)
print 'Promedio de No de comparaciones practico: ' , np.mean(comp)
print 'Peor No de comparaciones practico: ' , np.amax(comp)
print''
print'EVALUACION DE PREGUNTAS'
print 'Mejor No de preguntas practico: ' , np.amin(ques)
print 'Promedio de No de preguntas practico: ' , np.mean(ques)
print 'Peor No de preguntas practico: ' , np.amax(ques)
print''
print'EVALUACION DE INTERCAMBIOS'
print 'Mejor No de intercambios practico: ' , np.amin(intc)
print 'Promedio de No de intercambios practico: ' , np.mean(intc)
print 'Peor No de intercambios practico: ' , np.amax(intc)

plt.figure(1)
plt.subplot(221)
plt.hist(stv)
plt.title('Distribucion del numero de instrucciones')

plt.subplot(222)
plt.hist(comp)
plt.title('Distribucion del numero de comparaciones')

plt.subplot(223)
plt.hist(ques)
plt.title('Distribucion del numero de preguntas')

plt.subplot(224)
plt.hist(intc)
plt.title('Distribucion del numero de intercambios')


plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=0.35)

plt.show()

```
